## The Dataset
Our group will be using the [Emotion Detection](https://www.kaggle.com/ananthu017/emotion-detection-fer) dataset, a large image dataset from Kaggle containing 35,685 examples of 48x48 pixel gray scale images of faces divided into train and test dataset. 

We decided to choose this dataset because of the associated labels for each picture, and the large amount of images available. Many other datasets either did not have a large amount of images, or had no labels associated with the images.

```python
#Imports
import os
import tensorflow as tf
from tensorflow import keras
from matplotlib import pyplot as plt
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

## Sample of images in our dataset
Lets look at some of the images in each of our labels:

```python
folder_path = 'data/train/'
test_folder = 'data/test/'

def display_images(emotion):
    plt.figure(figsize= (6,6))
    for i in range(1, 10, 1):
        plt.subplot(3,3,i)
        img = image.load_img(folder_path + '/' + emotion + "/" + os.listdir(folder_path + '/' + emotion)[i], target_size=(48, 48))
        plt.title(emotion)
        plt.imshow(img)   
    plt.show()

for i in os.listdir('data/train'):
    display_images(i)
```

## Splitting the data
The image dataset is split between test and train data, with ~28,000 images in our training set and ~7000 images in our test set.

```python
# create a data generator
datagen = ImageDataGenerator()
# load training images
train = datagen.flow_from_directory('data/train/', class_mode='binary', batch_size=64)
# load test images
test = datagen.flow_from_directory('data/test/', class_mode='binary', batch_size=64)
```
